```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(12, 8))
plt.plot(x, y, 'rx-', label='sin(x)')
plt.plot(x, z, 'b+--', label='cos(x)')
plt.legend()

plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
A = np.random.random((15, 10))

plt.figure()

plt.subplot(1,3,1)
plt.imshow(A)

plt.subplot(1,3,2)
plt.imshow(A, origin='lower')

plt.subplot(1,3,3)
plt.pcolor(A)
```

```python
a = np.array([[10], [20], [30]])
b = np.array([[1, 2, 3, 4, 5]])
print("a=" + str(a))
print("b=" + str(b))
```

```python
# Use * for component wise multiplication
c = b * b
print("c = " + str(c))
```

```python
# Use @ for matrix multiplication
c = a @ b
print("C = a*b = " + str(c))
```

# Special matrices

```python
A = np.zeros((15, 10))
plt.figure()
plt.imshow(A, vmin=0, vmax=1)
plt.colorbar()
```

```python
A = np.ones((15, 10))
plt.figure()
plt.imshow(A, vmin=0, vmax=1)
plt.colorbar()
```

```python
A = np.eye(15, 10)
plt.figure()
plt.imshow(A, vmin=0, vmax=1)
plt.colorbar()
```

```python
A = np.random.random((15, 10))
plt.figure()
plt.imshow(A, vmin=0, vmax=1)
plt.colorbar()
```

# Indexing

```python
A = np.random.random((15, 10))
plt.figure()

plt.subplot(1,3,1)
plt.imshow(A, vmin=0, vmax=1)

plt.subplot(1,3,2)
plt.imshow(A[0:2,:], vmin=0, vmax=1)

plt.subplot(1,3,3)
plt.imshow(A[:,1:5], vmin=0, vmax=1)
```
