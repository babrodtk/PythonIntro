```python
# Importing packages

import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 1000

x = np.random.random(n)
y = np.random.random(n)
```

```python
fig = plt.figure(figsize=(18,12))
plt.plot(x, y, 'rx')
plt.axis('image')
```

```python
inside = np.where((x**2 + y**2) <= 1, True, False)
outside = ~inside
```

```python
fig = plt.figure(figsize=(18,12))
plt.plot(x[inside], y[inside], 'bx', label='Inside')
plt.plot(x[outside], y[outside], 'rx', label='Outside')
plt.axis('image')

#plot circle
x_c = np.linspace(0, 1, 100)
y_c = np.sqrt(1 - x_c**2)
plt.plot(x_c, y_c, '--')
```

```python
num_inside = np.sum(inside)
print(num_inside)
print(n)

#A = pi*r^2
pi = 4*(num_inside/n)
print(pi)
```
